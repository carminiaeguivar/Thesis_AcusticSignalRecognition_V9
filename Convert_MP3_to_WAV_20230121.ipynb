```python
import time
import os
from pydub import AudioSegment
MP3_SRC_DIR =  "/content/drive/MyDrive/Thesis/DataSets/BirdsBolivia/recordings_mp3/"
WAV_DEST_DIR = "/content/drive/MyDrive/Thesis/DataSets/BirdsBolivia/recordings_wav/"
start1 = time.time()
cont = 0
for clase in os.listdir(MP3_SRC_DIR):
    #audio_path =  MP3_SRC_DIR + name
    print(clase)     

    if int(clase) >= 17:   ### Concertimos a WAV las 10 primeras clases
        for fname in os.listdir(MP3_SRC_DIR + clase +"/"):
            src = MP3_SRC_DIR + clase +"/"+ fname
            label = fname.split(".")[0]
            #print(label)
            dst = WAV_DEST_DIR+ clase +"/"+label+".wav"
            print("soure:" +src)
            print("destiny:" +dst)
            #####################################################
            #cont = cont +1
            #if cont > 10:
            #    break    # break here
            #    print('Test to break ' + clase)
            #####################################################
            ## convertir to WAV
            #####################################################
            directorio  = WAV_DEST_DIR+clase+"/"
            #directorio = clase
            if not os.path.exists(directorio):  ## crear el folder para  la clase
                   os.mkdir(directorio)
            #convert
            audSeg = AudioSegment.from_mp3(src)
            audSeg.export(dst, format="wav")
end1 = time.time()
wav_time = end1-start1
print("----------------- Generating WAV files with : from pydub import AudioSegment  ---------------")
print("Completed Time: {:0.2f} minute".format(wav_time/60.0))
#print('Plotting examples:')        
```
