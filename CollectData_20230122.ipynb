```python
MP3_DIR = "https://cdn.download.ams.birds.cornell.edu/api/v1/asset/"
DATASET_DIR = "/content/drive/MyDrive/Thesis/DataSets/BirdsBolivia/recordings_mp3/"
```

```python
import requests
import os

####################################################################
# 1  read de excel file
####################################################################
for  f in flist:
    namef =  f + ".xls"
    df = pd.read_excel( namef )
####################################################################
# 2  read every audio in the file
####################################################################
    for ind in df.index:  ## list de  audios por archivo
        #print(df['catalogo'][ind], df['catalogo'][ind])
        audio_url = MP3_DIR + str(df['ML Catalog Number'][ind]) +'/'
        file_name = str(df['bird_code'][ind])+"_"+str(df['ML Catalog Number'][ind]) +"_"+str(df['num'][ind]) + ".mp3"
        print(audio_url + " : " + file_name)

        directorio = DATASET_DIR + str(df['bird_code'][ind])
        if not os.path.exists(directorio):
                    os.mkdir(directorio)
        ##############################################    
        #if df['num'][ind] == 10:
        #    break    # break here
        #    print('Number is ' + str(df['num'][ind])) 
        ###############################################       
        # URL of the audio to be downloaded is defined as image_url
        r = requests.get(audio_url) # create HTTP response object
        # send a HTTP request to the server and save
        # the HTTP response in a response object called r    
        with open(directorio + "/" + file_name,'wb') as f:
          
            # Saving received content as a png file in
            # binary format
          
            # write the contents of the response (r.content)
            # to a new file in binary mode.
            f.write(r.content)
```

```python
import time
import os
from pydub import AudioSegment
MP3_SRC_DIR =  "/content/drive/MyDrive/Thesis/DataSets/BirdsBolivia/recordings_mp3/"
WAV_DEST_DIR = "/content/drive/MyDrive/Thesis/DataSets/BirdsBolivia/recordings_wav/"
start1 = time.time()
for name in os.listdir(MP3_SRC_DIR):
    #audio_path =  MP3_SRC_DIR + name
    ##print(name)
    src = MP3_SRC_DIR + name
    label = name.split(".")[0]
    #print(label)
    dst = WAV_DEST_DIR+label+".wav"
    print(src)
    #print(dst)
    audSeg = AudioSegment.from_mp3(src)
    audSeg.export(dst, format="wav")
end1 = time.time()
wav_time = end1-start1
print("----------------- Generating WAV files with : from pydub import AudioSegment  ---------------")
print("Completed Time: {:0.2f} minute".format(wav_time/60.0))
#print('Plotting examples:')        
```
